## Desfasagem entre dois Sinais

```python
theta=np.linspace(0, 3*np.pi, 100) 

fig = plt.figure()
ax1 = fig.add_subplot(211, autoscale_on=False, xlim=(0, 10), ylim=(-1, 1))
ax2 = fig.add_subplot(212, autoscale_on=False, xlim=(-1, 1), ylim=(-1, 1))
#ax2.xaxis.label='x'

def plotPhase(delta):
    y=np.sin(theta)
    ax1.cla()
    ax1.grid()
    ax1.plot(theta, y,label='Emitido-CH1')
    y=np.sin(theta+delta)
    ax1.plot(theta, y, label='Recebido-CH2')
    ax1.legend(loc='upper center')
    x=np.sin(theta)
    y=np.sin(theta+delta)
    ax2.cla()
    ax2.grid()
    ax2.plot(x, y)
    #ax2.axis('equal')
    fig.show()

#ax1.sh
#plotPhase(np.pi/4)
widgets.interact(plotPhase, delta=(0,np.pi,0.01));
```
